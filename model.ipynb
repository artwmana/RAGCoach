```python
import psycopg2
```

```python
# Настройки подключения (поставь свои!)
HOST = "localhost"
PORT = 5432
USER = "admin"       # или postgres
PASSWORD = "secret"  # твой пароль
DATABASE = "exam_bot"

def main():
    try:
        conn = psycopg2.connect(
            host=HOST,
            port=PORT,
            user=USER,
            password=PASSWORD,
            dbname=DATABASE
        )
        conn.autocommit = True
        cur = conn.cursor()

        # Создаем таблицу
        cur.execute("""
            CREATE TABLE IF NOT EXISTS students (
                id SERIAL PRIMARY KEY,
                student_id VARCHAR(50) UNIQUE NOT NULL,
                name VARCHAR(255) NOT NULL,
                group_name int,
                created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
            );
        """)

        # Добавляем данные
        cur.execute("""
            INSERT INTO students (student_id, name, group_name)
            VALUES
                ('S001', 'Artem Morozov', '21'),
                ('S002', 'Ivan Petrov', '22')
            ON CONFLICT (student_id) DO NOTHING;
        """)

        print("Таблица создана и данные добавлены!")

    except Exception as e:
        print("Ошибка:", e)

    finally:
        if conn:
            cur.close()
            conn.close()

if __name__ == "__main__":
    main()
```
