```python
import json
import random
import re
from pathlib import Path
from ragcoach import pdf_to_json, build_grader
from ragcoach.infrastructure.db import QdrantService

from pathlib import Path
import json
from sentence_transformers import SentenceTransformer

from qdrant_client import QdrantClient
from qdrant_client.models import VectorParams, Distance, PointStruct
import uuid
from tqdm import tqdm
```

```python
include_context = True
make_json_dir = False # Если json/ пуст 

path = Path("/Users/artem/Documents/Projects/RAGCoach")

pdf_dir = path.joinpath("data/lectures")
json_dir = path.joinpath("data/json")
```

```python
svc = QdrantService(chunk_words=400)

if make_json_dir:
    for pdf_path in sorted(pdf_dir.glob("*.pdf")):
        json_path = json_dir / f"{pdf_path.stem}.json"
        pdf_to_json(str(pdf_path), str(json_path))
        svc.ingest_json(json_path, source_name=pdf_path.stem)
```

```python
questions_raw = [q for q in (path.joinpath("data/questions.txt")).read_text(encoding='utf-8').splitlines() if q.strip()]
question_line = random.choice(questions_raw)
question_text = re.sub(r'^\s*\d+\s*[\.|\)]?\s*', '', question_line).strip()

print(question_text)
```

```python
hits = svc.search(question_text, top_k=5)
context = "".join(h["payload"].get("text", "") for h in hits)
print(context)
```

```python
sample_answer = input()
```

```python
import asyncio
from ragcoach.infrastructure.llm import OllamaLLMGateway

llm = OllamaLLMGateway()
answer = await llm.generate(question_text, context, sample_answer)
print(answer)
```
