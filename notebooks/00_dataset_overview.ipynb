```python
import psycopg2
```

```python
# Настройки подключения (поставь свои!)
HOST = "localhost"
PORT = 5432
USER = "admin"       # или postgres
PASSWORD = "secret"  # твой пароль
DATABASE = "exam_bot"

conn = psycopg2.connect(
        host=HOST,
        port=PORT,
        user=USER,
        password=PASSWORD,
        dbname=DATABASE
    )
conn.autocommit = True
cur = conn.cursor()

cur.execute("""
            SELECT id, student_id, name, group_name, created_at
            FROM public.students;
        """)
```

```python
rows = cur.fetchall()
for row in rows:
    print(row)
```
